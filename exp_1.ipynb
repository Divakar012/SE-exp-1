```python

import math
import numpy as np
import matplotlib.pyplot as plt

def quadratic_solver(a, b, c):
    discriminant = b**2 - 4*a*c
    if a == 0:
        print("Error: Coefficient 'a' cannot be zero.")
        return None, None, False
    if discriminant > 0:
        root1 = (-b + math.sqrt(discriminant)) / (2 * a)
        root2 = (-b - math.sqrt(discriminant)) / (2 * a)
        return root1, root2, True
    elif discriminant == 0:
        root = -b / (2 * a)
        return root, root, True
    else:
        return None, None, False

def plot_graph(a, b, c):
    x = np.linspace(-10, 10, 500)
    y = a * x**2 + b * x + c
    plt.plot(x, y, label=f"{a}xÂ² + {b}x + {c}")
    plt.axhline(0, color='black', linestyle='--', linewidth=0.8)
    plt.axvline(0, color='black', linestyle='--', linewidth=0.8)
    plt.title("Graph of the Quadratic Equation")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid()
    plt.show()

def main():
    try:
        a = float(input("Enter coefficient a: "))
        b = float(input("Enter coefficient b: "))
        c = float(input("Enter coefficient c: "))
        root1, root2, has_roots = quadratic_solver(a, b, c)
        if has_roots:
            print(f"Roots: {root1:.2f}, {root2:.2f}")
        else:
            print("No real roots.")
        plot_graph(a, b, c)
    except ValueError:
        print("Invalid input. Please enter numeric values.")

main()
```
